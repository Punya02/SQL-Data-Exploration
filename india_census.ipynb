```python
#num of state , count on num of occurenece , sorting value

census_data['State_name'].value_counts().sort_values().tail(10)
```

```python
#plotting population vs state

plt.figure(figsize = (12,8))
sns.set_theme(style="whitegrid")
census_data['State_name'].value_counts().plot(kind = 'barh')
```

```python
#proportions of state in %

census_data.State_name.value_counts(normalize = True).head()
```

```python
# filterring all records from state 'GOA' , 'LAKSHADWEEP' , 'CHANDIGARH'

state = census_data[census_data['State_name'].isin(['GOA' , 'LAKSHADWEEP' , 'CHANDIGARH'  ])]
state
```

```python
# sorting data with the sum of population in each state

census_data.groupby('State_name').Population.sum().sort_values(ascending =  False).head()
```

```python
# filterring all records from district 'New Delhi' , 'Lucknow' , 'Jaipur'

district = census_data[census_data['District_name'].isin([ 'New Delhi' , 'Lucknow' , 'Jaipur'])]
district
```
